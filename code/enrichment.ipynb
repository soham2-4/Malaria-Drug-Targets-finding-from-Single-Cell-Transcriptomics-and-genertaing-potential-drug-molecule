```python
import pandas as pd
df = pd.read_csv('/content/gProfiler_pfalciparum_05-01-2024_20-50-38__intersections.csv')
```

```python
df
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns  # Optional, for additional color palette options

# Assuming your DataFrame is named 'df'
# If 'df' is not defined, you should read your CSV file into a DataFrame using pd.read_csv

# Increase the figure size
fig, ax = plt.subplots(figsize=(12, 6))

# Counter for total bars
total_bars = 0

# Create a color palette with seaborn (optional, you can choose your own colors)
colors = sns.color_palette("husl", n_colors=len(df['source'].unique()))

# Iterate through each row and plot a bar
for index, row in df.iterrows():
    # Map each unique 'source' value to a color
    color = colors[df['source'].unique().tolist().index(row['source'])]

    # Plot a bar for each row
    ax.bar(str(index), row['negative_log10_of_adjusted_p_value'], color=color)

    # Increment the counter
    total_bars += 1

# Customize the plot
ax.set_xlabel('Term Names')
ax.set_ylabel('negative_log10_of_adjusted_p_value')
#ax.set_title('Bar Chart of Values for Each Term with Source')

# Set x-axis labels
ax.set_xticks(range(len(df)))
ax.set_xticklabels(df['term_name'], rotation='vertical')

# Create a legend for colors
handles = [plt.Rectangle((0, 0), 1, 1, color=color) for color in colors]
labels = df['source'].unique().tolist()
ax.legend(handles, labels, title='Source', loc='upper right')

# Show the plot
plt.show()

# Display the total number of bars
print(f'Total number of bars: {total_bars}')
```
