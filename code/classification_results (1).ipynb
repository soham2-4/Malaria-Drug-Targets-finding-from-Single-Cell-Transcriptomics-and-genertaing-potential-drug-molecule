```python
import numpy as np
import matplotlib.pyplot as plt

def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        plt.text(rect.get_x() + rect.get_width()/2., height,
                 '%.3f' % height,
                 ha='center', va='bottom')

X = ['Logistic_Regression', 'Random_Forest', 'XG_boost']

Logistic_Regression = [.948, .949, .423]
Random_Forest = [.946, .950, .408]
XG_boost = [.952, .952, .420]

Without_feature_selection = [Logistic_Regression[0], Random_Forest[0], XG_boost[0]]
With_feature_selection = [Logistic_Regression[1], Random_Forest[1], XG_boost[1]]
Randomly_selected_feature = [Logistic_Regression[2], Random_Forest[2], XG_boost[2]]

X_axis = np.arange(len(X))
bar_width = 0.25  # Adjust the width of the bars

bar1 = plt.bar(X_axis, Without_feature_selection, bar_width, label='Without_feature_selection')
bar2 = plt.bar(X_axis + bar_width, With_feature_selection, bar_width, label='With_feature_selection')
bar3 = plt.bar(X_axis + 2 * bar_width, Randomly_selected_feature, bar_width, label='Randomly_selected_feature')

plt.xticks(X_axis + bar_width, X)  # Adjust the position of x-axis ticks
#plt.xlabel("Feature Selection Method")
plt.ylabel("Accuracy")
#plt.title("Comparison between Different Feature Selection Methods")
plt.legend()

# Add labels on top of the bars
autolabel(bar1)
autolabel(bar2)
autolabel(bar3)

# Adjust y-axis limits for more space
plt.ylim(0, 1.5)  # Increase the upper limit

plt.show()
```
